```python
import numpy as np
file_face = np.load('ds_colorferet_emb.npy')
file_age = np.load('ds_colorferet_age.npy')
file_eth = np.load('ds_colorferet_ethnicity.npy',allow_pickle=True)
file_gender = np.load('ds_colorferet_gender.npy',allow_pickle=True)
```

```python
import pandas as pd
from sklearn.manifold import TSNE
emb=TSNE(n_components=2, learning_rate='auto', init='random', perplexity=3).fit_transform(file_face)

tsne_data = np.vstack((emb.T, file_age)).T
tsne_df = pd.DataFrame(data = tsne_data,
     columns =("Dim_1", "Dim_2", "label"))
```
